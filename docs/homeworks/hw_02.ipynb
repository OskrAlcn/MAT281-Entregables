```python
import matplotlib.dates as mdates
from matplotlib import ticker

# ========================
# 1) Preparar los datos
# ========================
# Asegurar que 'time' es datetime
merged['time'] = pd.to_datetime(merged['time'])

# Construir temp igual que el código original
temp = pd.DataFrame(merged.groupby(['time', 'id'])['time'].count())
temp = temp.rename(columns={"time": "amount"})
temp = temp.reset_index(drop=False)
temp = temp.groupby('time')['amount'].mean()

# Índice de fechas
dates = pd.to_datetime(temp.index)
y = temp.values
ymax = y.max()

# Convertir fechas a números (días) para matplotlib
x = mdates.date2num(dates)

# Rango de verano en números
summer_dates = pd.date_range("2020-06-01", "2020-08-23", freq="D")
summer = mdates.date2num(summer_dates)

# Fechas importantes en números
cierre_escuelas = mdates.date2num(np.datetime64("2020-02-12"))
covid_pandemia = mdates.date2num(np.datetime64("2020-03-11"))
vacaciones_dia = mdates.date2num(np.datetime64("2020-06-27"))

# ========================
# 2) Gráfico
# ========================
background_color = "#B4DBE9"

plt.rcParams['figure.dpi'] = 300
fig = plt.figure(figsize=(9, 2), facecolor=background_color)   # un poco más alto
gs = fig.add_gridspec(1, 1)
gs.update(wspace=0, hspace=0)
ax0 = fig.add_subplot(gs[0, 0])
ax0.set_facecolor(background_color)

# --- barras ---
ax0.bar(
    x, y,
    width=1.0,
    color='white',        # barras blancas
    edgecolor='dimgray',  # bordes grises
    linewidth=0.5,
    zorder=2
)

# --- franja amarilla de vacaciones de verano ---
ax0.fill_between(
    summer,
    0,
    ymax,
    color='#ffd514',
    alpha=0.5,
    zorder=1,
    linewidth=0
)

# --- líneas verticales ---
ax0.axvline(cierre_escuelas, color='#ffd514', alpha=0.7)
ax0.axvline(covid_pandemia, color='#ffd514', alpha=0.7)

for s in ["right", "top"]:
    ax0.spines[s].set_visible(False)
ax0.grid(which='major', axis='x', zorder=0, color='#EEEEEE', lw=0.3)
ax0.grid(which='major', axis='y', zorder=0, color='#EEEEEE', lw=0.3)

# limitar el eje X de enero a diciembre 2020
ax0.set_xlim(
    mdates.date2num(np.datetime64("2020-01-01")),
    mdates.date2num(np.datetime64("2020-12-31"))
)

# === títulos dentro del área ===
x0, x1 = ax0.get_xlim()
y0, y1 = ax0.get_ylim()

ax0.text(
    x0, y1*0.97,
    'Promedio diario de productos accedidos',
    color='black', fontsize=7,
    ha='left', va='top', weight='bold'
)
ax0.text(
    x0, y1*0.90,
    'Después de las vacaciones de verano, hay un aumento en los productos accedidos',
    color='#292929', fontsize=5,
    ha='left', va='top'
)

ax0.annotate(
    "Cierre temporal de escuelas.",
    xy=(cierre_escuelas, ymax*0.8),
    xytext=(mdates.date2num(np.datetime64("2020-01-10")), ymax*0.65),
    arrowprops=dict(arrowstyle="->"),
    fontsize=5
)
ax0.annotate(
    "COVID-19\nPandemia",
    xy=(covid_pandemia, ymax*0.8),
    xytext=(mdates.date2num(np.datetime64("2020-03-20")), ymax*0.6),
    arrowprops=dict(arrowstyle="->"),
    fontsize=5
)
ax0.annotate(
    "Vacaciones de verano",
    xy=(vacaciones_dia, ymax*0.4),
    xytext=(vacaciones_dia, ymax*0.4),
    fontsize=5
)

# === ejes ===
ax0.set_xlabel("fecha", fontsize=6, weight='bold')
ax0.set_ylabel("productos", fontsize=6, weight='bold')

ax0.tick_params('both', length=2, which='major', labelsize=5)
ax0.xaxis.set_major_locator(mdates.MonthLocator())
ax0.xaxis.set_major_formatter(mdates.DateFormatter('%b 2020'))  # Jan 2020, Feb 2020, ...

y_format = ticker.FuncFormatter(lambda x, p: format(int(x), ','))
ax0.yaxis.set_major_formatter(y_format)

plt.tight_layout()
plt.show()
```

Cada mes presenta fluctuaciones, que sigue un patrón de 5-2, es decir, 5 días de escuela y 2 días de fin de semana. Aunque el número de productos accedidos disminuye durante el fin de semana, aún observamos productos accedidos en esos días, siendo aproximadamente un 25%-50% en comparación con los días de la semana. Esto tiene sentido, ya que los fines de semana no son días de clases.

### Comentario:

El dataframe contiene **891 datos** y **12 columnas**, el target (`'Survived'`) y los atributos, tipo numéricos y categóricos.  

Todos los datos están etiquetados, es decir, la columna `'Survived'` (target) no presenta datos nulos.  

Algunas columnas poseen datos nulos. Estas son:

- **Age** (numérica): 714 datos no nulos.  
- **Cabin** (categórica): 204 datos no nulos.  
- **Embarked** (categórica): 889 datos no nulos.  

Dado que la columna `'Cabin'` posee más datos nulos que no nulos, eliminaremos esta columna para no perturbar tanto la muestra.  

Para la variable **'Age'**, reemplazaremos los datos nulos con la **desviación estándar**.  

Para la variable **'Embarked'**, reemplazaremos con la **moda**.

```python
df_train2 = df_train.copy()

df_train2 = df_train2.drop('Cabin', axis=1)

std_age = df_train2['Age'].std()
df_train2['Age'] = df_train2['Age'].fillna(std_age)

moda_embarked = df_train2['Embarked'].mode().iloc[0]
df_train2['Embarked'] = df_train2['Embarked'].fillna(moda_embarked)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_num_scaled = scaler.fit_transform(df_num)
df_num_scaled = pd.DataFrame(df_num_scaled, columns=df_num.columns, index = X_train.index)
print('Datos numéricos estandarizados:')
display(df_num_scaled.head())
```

```python
from sklearn.feature_selection import f_classif

scores = []
for col in df_num.columns.values:
  x = df_num_scaled[col].values.reshape(-1, 1)
  score = f_classif(x, y_train)[0][0]
  scores.append(score)
```

```python
fig, axes = plt.subplots(figsize=(6,4), nrows=1, ncols=1)
plot_data = pd.DataFrame({'atributos_num': df_num.columns.values, 'fisher_score': scores})
plot_data = plot_data.sort_values(by='fisher_score', ascending=False)
p = sns.barplot(data=plot_data, x='fisher_score', y='atributos_num', palette="Blues_d",ax=axes)
plt.show()
```

**Comentario:** Los atributos numéricos más relevantes son 'Pclass' y 'Fare', basandonos en la puntuación fisher-score entre cada una de estas con el target.

```python
from sklearn.feature_selection import mutual_info_classif as mi
from sklearn.preprocessing import LabelEncoder

scores = []
for col in df_categ.columns.values:
  x = LabelEncoder().fit_transform(df_train2[col]).reshape(-1,1)
  score = mi(x, y_train, discrete_features=True)[0]
  scores.append(score)
```

```python
fig, axes = plt.subplots(figsize=(4,4), nrows=1, ncols=1)
plot_data = pd.DataFrame({'x': df_categ.columns.values, 'y': scores})
plot_data = plot_data.sort_values(by='y', ascending=False)
p = sns.barplot(data=plot_data,x='y', y='x', palette="Blues_d",ax=axes)
plt.show()
```



**Comentarios:** Tal como se intuía a partir de la visualización, el atributo 'Sex' es muy relevante para la predicción, a diferencia de 'Embarked'. Esto, en base a la información mutua entre cada atributo con el target.

 La desventaja de recortar variables en base a la puntuación fisher-score y de la información mutua es que se tratan de métodos de filtrado individual, es decir, no tienen la capacidad de detectar redundancias ni complementariedades entre atributos.

### Grid Search: Logistic Regressor

```python
# Parámetros:
param_grid_lr = {'penalty' : ['l1', 'l2'],
                 'C': [0.001, 0.01, 0.1, 1, 10, 100],
                 'class_weight' : ['balanced', None]
                 }
param_grid_lr

# Aplicar greed search:
gs_lr = GridSearchCV(estimator = lr,
                  param_grid = param_grid_lr,
                  scoring = 'accuracy',
                  cv = 5,
                  n_jobs = -1)

gs_lr = gs_lr.fit(X_train_procesado, y_train)

# Imprimir resultados:
print(gs_lr.best_score_)
print(gs_lr.best_params_)
```

```python
# Parámetros:
param_grid_svc = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'gamma': ['scale', 'auto'],
    'class_weight' : ['balanced', None]
}

# Aplicar greed search:
gs_svc = GridSearchCV(estimator = svc,
                  param_grid = param_grid_svc,
                  scoring = 'accuracy',
                  cv = 5,
                  n_jobs = -1)

gs_svc = gs_svc.fit(X_train_procesado, y_train)

# Imprimir resultados:
print(gs_svc.best_score_)
print(gs_svc.best_params_)
```

```python
# Parámetros:
param_grid_rfc = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'class_weight' : ['balanced', None]
}

# Aplicar greed search:
gs_rfc = GridSearchCV(estimator = rfc,
                  param_grid = param_grid_rfc,
                  scoring = 'accuracy',
                  cv = 5,
                  n_jobs = -1)

gs_rfc = gs_rfc.fit(X_train_procesado, y_train)

# Imprimir resultados:
print(gs_rfc.best_score_)
print(gs_rfc.best_params_)
```
